```python
import gym
import numpy as np
import random
import math
```

```python
env = gym.make('CartPole-v0')
learningrate = 0.5
discountrate = 0.8 # discount
episodes = 200
```

```python
class Agent():
    
    def __init__(self, env):
        self.probability = 1
        self.episodenum = 0
        self.action_size = env.action_space.n
        testlist = [-4.8,-5,-24,-5]
        largelist = []
        self.tablelist = []
        for i in range(960):   
            for p in np.arange(-4.80,4.80,0.01):
                p = format(p,'.2f')
                largelist.append(float(p))
            carpos = largelist[i]
            for j in range(11):
                carvelo = testlist[1]+1*j
                for k in range(49):
                    poleang = testlist[2]+1*k
                    for h in range(11):
                        polevelo = testlist[3]+1*h
                        self.tablelist.append([carpos,carvelo,poleang,polevelo])           
        self.rlList = [[0]*2]*len(self.tablelist)    
        
    
    def chooseStep(self, state, reward, action, paststate):
        #state[0] is the only float, multiply by 10,math.floor it, then divide by 10.
#         x = state[0]*10
#         x = (math.floor(x))/10
        x = state[0]
        x = float(format(x,'.2f'))
        y = state[1]
        y = math.floor(y)
        z = state[2]
        z = math.floor(z)
        w = state[3]
        w = math.floor(w)
        self.newstate = [x,y,z,w]
        print(self.newstate)
        index = self.tablelist.index(self.newstate)

        if self.rlList[index][0]>self.rlList[index][1]: choice = 0
        elif self.rlList[index][0]<self.rlList[index][1]: choice = 1 
        else:
            choice = random.choice([0,1])
        
        if reward != -1:
#             q = paststate[0]*10
#             q = (math.floor(q))/10
            q = paststate[0]
            q = format(q,'.2f')
            q = float(q)
            w = paststate[1]
            w = math.floor(w)
            e = paststate[2]
            e = math.floor(e)
            r = paststate[3]
            r = math.floor(r)
            paststate = [q,w,e,r]
        
            pastindex = self.tablelist.index(paststate)
            
            if self.rlList[index][0]>self.rlList[index][1]: maxnumrow = self.rlList[index][0]
            elif self.rlList[index][0]<self.rlList[index][1]: maxnumrow = self.rlList[index][1]
            else: maxnumrow = self.rlList[index][random.choice([0,1])]
            qnum = (1 - learningrate) * self.rlList[pastindex][action] + learningrate*(reward+discountrate*(maxnumrow)) #bellman equation
            self.rlList[pastindex][action] = qnum  
            print(newagent.rlList[index])
            # determine next state of car, based on the direction it is moving now (if -velo moving left if +velo moving right)
            # determine best CURRENT action based on if next state is good (gives reward)
            # when choice = 0 (move left), -1 is added to velo, when choice = 1 (right), +1 velo is added.
            # paststate needed in order to know where to append reward (what index).
        
        if self.episodenum%100 == 0:
            self.probability += 1
        self.episodenum += 1
            

        explorerate = random.choice(range(1,self.probability)) # replace second number with self.probability
        if explorerate == 1:
            print("exploring!")
            return 1-choice
        else:
            return choice
        
    
newagent = Agent(env)
for p in range(episodes):
    state = env.reset()
    done = False
    action = newagent.chooseStep(state, -1, -1,0)
    while not done:
        paststate = state
        state, reward, done, info = env.step(action)
        action = newagent.chooseStep(state,reward,action,paststate)
        env.render()
    env.close()
    
```
